```python
import pandas as pd
```

```python
country_codes = pd.read_csv('./data/country_codes.csv')
rankings = pd.read_csv('./data/fifa_rankings.csv')
matches = pd.read_csv('./data/matches.csv')
```

```python
matches
```

```python
rankings['rank_date'].iloc[0]
```

```python
matches_trimmed = matches.copy()
```

```python
# Trim matches
matches_trimmed.drop(matches_trimmed[matches_trimmed['date'] <= rankings['rank_date'].iloc[0]].index, inplace = True)
matches_trimmed
```

```python
# Countries in Rankings but NOT in Matches


# first see which countries are in rankings but not in matches
countries_in_rankings_not_in_matches = list(set(rankings['country_full'].unique()) - set(matches_trimmed['home_team'].unique()))

# Create series with all countries in matches and the number of matches
countries_list_matches = rankings['country_full'].value_counts()

# Then filter the  list of matches using the countries that are in matches but not in federation
print((countries_list_matches.loc[countries_in_rankings_not_in_matches]).sort_values(ascending=False))

# This way we see how we should reformat the countries from confederations
```

```python
# Countries IN Matches but NOT IN Rankings


# first see which countries are in matches but not in rankings
countries_in_matches_not_in_rankings = list(set(matches_trimmed['home_team'].unique()) - set(rankings['country_full'].unique()))

# Create series with all countries in matches and the number of matches
countries_list = matches_trimmed['home_team'].value_counts()

# Then filter the list of matches using the countries that are in matches but not in rankings (first 25 with most matches)
print((countries_list.loc[countries_in_matches_not_in_rankings]).sort_values(ascending=False).head(25))
```
