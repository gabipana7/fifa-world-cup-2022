```python
import pandas as pd
```

```python
country_codes = pd.read_csv('./data/country_codes.csv')
rankings = pd.read_csv('./data/fifa_rankings.csv')
matches = pd.read_csv('./data/matches.csv')
```

```python
# Trim matches
matches_trimmed = matches.drop(matches[matches['date'] <= rankings['rank_date'].iloc[0]].index)
matches_trimmed
```

```python
# Countries in Rankings but NOT in Matches


# first see which countries are in rankings but not in matches
countries_in_rankings_not_in_matches = list(set(rankings['team'].unique()) - set(matches_trimmed['home_team'].unique()))
countries_in_rankings_not_in_matches

# # Create series with all countries in matches and the number of matches
# countries_list_matches = rankings['team'].value_counts()

# # Then filter the  list of matches using the countries that are in matches but not in federation
# print((countries_list_matches.loc[countries_in_rankings_not_in_matches]).sort_values(ascending=False))
```

```python
list(set(rankings['team'].unique()) - set(matches_trimmed['away_team'].unique()))
```

### 3 Countries that are in RANKINGS do not show up in MATCHES
### Because they have different names in MATCHES

```python
# Countries IN Matches but NOT IN RANKINGS
# We are looking for China, Taiwan (Chinese Taipei) and Unites States Virgin Islands (U.S. Virgin Islands)
# These are popular countries, they will for sure have more matches to them

# first see which countries are in matches but not in rankings (there will be more than 3, MATCHES has non-fifa countries as well)
countries_in_matches_not_in_rankings = list(set(matches_trimmed['home_team'].unique()) - set(rankings['team'].unique()))

# Create series with all countries in MATCHES and the number of matches
countries_list = matches_trimmed['home_team'].value_counts()

# Then filter the list of matches using the countries that are in MATCHES but not in RANKINGS (first 25 with most matches)
print((countries_list.loc[countries_in_matches_not_in_rankings]).sort_values(ascending=False).head(20))
```

```python
teams_to_replace = ['United States Virgin Islands','China PR','Taiwan']
print('Teams to replace: \n',teams_to_replace, '\nTo be replaced by: \n',countries_in_rankings_not_in_matches)
```

```python
matches_trimmed_modif = matches_trimmed.copy()
```

```python
matches_trimmed_modif['home_team'] = matches_trimmed['home_team'].replace(teams_to_replace,countries_in_rankings_not_in_matches)
matches_trimmed_modif['away_team'] = matches_trimmed['away_team'].replace(teams_to_replace,countries_in_rankings_not_in_matches)
matches_trimmed_modif['country'] = matches_trimmed['country'].replace(teams_to_replace,countries_in_rankings_not_in_matches)
```

```python
matches_trimmed_modif
```

```python
# test
list(set(rankings['team'].unique()) - set(matches_trimmed_modif['home_team'].unique()))
# all teams from RANKINGS have their counterpart in MATCHES
```

```python
# But there are still non-fifa countries in matches as home_team
nonfifa_still_in_matches_home = list(set(matches_trimmed_modif['home_team'].unique()) - set(rankings['team'].unique()))
print('Non-fifa teams in international matches (as home_team): \n', nonfifa_still_in_matches_home)
```

```python
# But there are still non-fifa countries in matches as away_team
nonfifa_still_in_matches_away = list(set(matches_trimmed_modif['away_team'].unique()) - set(rankings['team'].unique()))
print('Non-fifa teams in international matches (as away_team): \n', nonfifa_still_in_matches_away)
```

```python
nonfifa_still_in_matches = list(set(nonfifa_still_in_matches_home+nonfifa_still_in_matches_away))
print('Total nonfifa countries in international matches: \n', nonfifa_still_in_matches)
```

```python
# We can delete all of these matches from the record so that only fifa matches are shown
fifa_matches = matches_trimmed_modif.copy()
```

```python
fifa_matches.drop(fifa_matches[fifa_matches['home_team'].isin(nonfifa_still_in_matches)].index, inplace=True)
```

```python
fifa_matches
```

```python
fifa_matches.drop(fifa_matches[fifa_matches['away_team'].isin(nonfifa_still_in_matches)].index, inplace=True)
```

```python
fifa_matches
```

```python
fifa_matches.drop(fifa_matches[fifa_matches['country'].isin(nonfifa_still_in_matches)].index, inplace=True)
```

```python
fifa_matches
```

```python
fifa_matches.reset_index(drop=True, inplace=True)
```

```python
fifa_matches.drop(columns='match_id', inplace=True)
```

```python
fifa_matches.insert(0,"match_id", fifa_matches.index+1)
```

```python
fifa_matches
```

```python
fifa_matches.to_csv('./data/fifa_matches.csv', index=False)
```
